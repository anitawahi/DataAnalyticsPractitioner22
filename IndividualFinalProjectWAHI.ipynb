```python
def grade_type(ans):
    if ans == "Exceptional-A":
        return 13
    if ans == "Exceptional-B":
        return 12
    if ans == "Exceptional-C":
        return 11
    if ans == "Exceptional-D":
        return 10
    if ans == "Superior":
        return 9
    if ans == "Excellent":
        return 8
    if ans == "Very Good":
        return 7
    if ans == "Above Average":
        return 6
    if ans == "Average":
        return 5
    if ans == "Good Quality":
        return 4
    if ans == "Fair Quality":
        return 3
    if ans == "Low Quality":
        return 2
    if ans == "No Data":
        return 1

df['GRADE1'] = df["GRADE"].apply(grade_type)
df["GRADE1"].value_counts()
```

```python
def cond_type(ans):
    if ans == "Excellent":
        return 7
    if ans == "Very Good":
        return 6
    if ans == "Average":
        return 5
    if ans == "Good":
        return 4
    if ans == "Fair":
        return 3
    if ans == "Poor":
        return 2
    if ans == "Default":
        return 1

df['CNDTN1'] = df["CNDTN"].apply(cond_type)
df["CNDTN1"].value_counts()
```

```python
# verifying that values align as previously
pd.crosstab(df["CNDTN1"], df["GRADE1"])
```

```python
df["WARD"].value_counts()
```

```python
def ward_num(ans):
    if ans == "Ward 8":
        return 8
    if ans == "Ward 7":
        return 7
    if ans == "Ward 6":
        return 6
    if ans == "Ward 5":
        return 5
    if ans == "Ward 4":
        return 4
    if ans == "Ward 3":
        return 3
    if ans == "Ward 2":
        return 2
    if ans == "Ward 1":
        return 1
    
df['WARD1'] = df["WARD"].apply(ward_num)
df["WARD1"].value_counts()
```

## 2. Removing Outliers and Inappropriate Values

```python
df2 = df.copy()
```

```python
# 1. rooms

# creating new column with z-scores
df2["zscoreROOMS"] = np.abs(stats.zscore(df2["ROOMS"]))

# identifying outliers with z-scores > 3
outlierROOMS = df2.loc[df2["zscoreROOMS"]>3]

print(outlierROOMS)

# 2,122 outliers have 15+ ROOMS (median is 6)

# removing outliers
df2.drop(df2.loc[df2["zscoreROOMS"]>3].index, inplace = True)
```
